## Plotting with Matplotlib

```python
# import nump and matplotlib
import numpy as np
import matplotlib.pyplot as plt
```

```python
# define the potential energy
def potential_energy(x, k=1.0):
    return 0.5*k*x**2 # k is the spring constant

# test function for the potental energy
potential_energy(2, 10)
```

```python
# define mesh for plotting
x = np.linspace(-10, 10, 1000) # np.linspace(x_min, x_max, npts)
y = potential_energy(x)

# testing the output
print(np.transpose([x, y]))
```

```python
# plot the function
plt.plot(x, y)

# add axes labels
plt.xlabel("displacement(arbitrary units)")
plt.ylabel("potential energy(arbitrary units)")

# change limits of x and y
plt.xlim((-4, 4))
plt.ylim((-1, 10))

plt.show() # combine all elements of the plot
```

## Contour Plots

```python
# import nump and matplotlib
import numpy as np
import matplotlib.pyplot as plt
```

```python
# define the potential energy
def potential_energy_2D(x, y, kx=1.0, ky=1.0):
    return 0.5*(kx*x**2 + ky*y**2) # (kx, ky) are the spring constants

# test function for the potental energy
potential_energy_2D(2, 1)
```

```python
# define meshes for plotting
x = np.linspace(-5, 5, 100) # np.linspace(xmin, xmax, npts)
y = np.linspace(-5, 5, 100) # np.linspace(ymin, ymax, npts)

# combine 1D meshes into meshgrid
X,Y = np.meshgrid(x, y)
# define z values
Z = potential_energy_2D(X,Y)

# testing the mesh
# print(X)
# print()
# print(Y)
# print()
# print(Z)
```

```python
# construct the contour plot
plt.figure(figsize=(4,3)) # figure size in inches
plt.contourf(X, Y, Z, levels=20)

# add a colorbar legend
plt.colorbar(label="potential energy (arbitrary units)")

# add axes labels
plt.xlabel("x-displacement(arbitrary units)")
plt.ylabel("y-displacement(arbitrary units)")

# change limites of x and y
plt.xlim((-4, 4))
plt.ylim((-4, 4))

plt.show() # combine all plot objects
```
