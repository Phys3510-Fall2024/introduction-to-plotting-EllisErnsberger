$U(Q_1,Q_2) = \frac{1}{2} K_1 Q_1^2 + \frac{1}{2} K_2 Q_2^2 + C Q_1 Q_2^2$

```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider, IntSlider # import the interact function and the sliders we need for this environment
```

```python
# Define the function to plot the first contour plot
def plot_contour1(k1=1, Q2=0, levels=10):
    # Define x, y range
    x_min, x_max = -5, 5
    y_min, y_max = -5, 5
    
    # Generate the grid
    q1 = np.linspace(x_min, x_max, 400)
    q2 = np.linspace(y_min, y_max, 400)
    Q1, Q2 = np.meshgrid(q1, q2)
    
    # Define the function U = k1 * Q1^2 + Q2
    Z = k1 * Q1**2 + Q2
    
    # Create the plot
    contour_lines = plt.contour(Q1, Q2, Z, levels=int(levels), colors='black', linewidths=0.5)
    contour_filled = plt.contourf(Q1, Q2, Z, levels=int(levels), cmap='Blues')
    plt.colorbar(contour_filled)
    
    # Set axis labels and range
    plt.xlabel('$q_1$', fontsize=12)
    plt.ylabel('$U$', fontsize=12)
    plt.title('Energy U vs. $q_1$', fontsize=14)
    plt.xlim([x_min, x_max])
    plt.ylim([y_min, y_max])

# Define the function to plot the second contour plot
def plot_contour2(k1=1, k2=1, c=0, levels=10):
    # Define x, y range
    x_min, x_max = -5, 5
    y_min, y_max = -5, 5
    
    # Generate the grid
    q1 = np.linspace(x_min, x_max, 400)
    q2 = np.linspace(y_min, y_max, 400)
    Q1, Q2 = np.meshgrid(q1, q2)
    
    # Define the function U = k1 * Q1^2 + k2 * Q2^2 + c * Q1 * Q2^2
    Z = k1 * Q1**2 + k2 * Q2**2 + c * Q1 * Q2**2
    
    # Create the plot
    contour_lines = plt.contour(Q1, Q2, Z, levels=int(levels), colors='black', linewidths=0.5)
    contour_filled = plt.contourf(Q1, Q2, Z, levels=int(levels), cmap='Oranges')
    plt.colorbar(contour_filled)
    
    # Set axis labels and range
    plt.xlabel('$q_1$', fontsize=12)
    plt.ylabel('$q_2$', fontsize=12)
    plt.title('Energy U vs. $q_1$ and $q_2$', fontsize=14)
    plt.xlim([x_min, x_max])
    plt.ylim([y_min, y_max])

# Define the function to plot both side-by-side
def plot_sidebyside(k1=1, k2=1, c=0, levels=10):
    plt.figure(figsize=(9, 4))
    
    # Plot the first contour on the left
    plt.subplot(1, 2, 1)
    plot_contour1(k1=k1, Q2=0, levels=levels)
    
    # Plot the second contour on the right
    plt.subplot(1, 2, 2)
    plot_contour2(k1=k1, k2=k2, c=c, levels=levels)
    
    # Adjust layout for better spacing
    plt.tight_layout()
    plt.show()

# Create interactive widgets to adjust the parameters
interact(plot_sidebyside,
         k1=FloatSlider(min=-2, max=2, step=0.1, value=1, description='k1'),
         k2=FloatSlider(min=-2, max=2, step=0.1, value=1, description='k2'),
         c=FloatSlider(min=-2, max=2, step=0.1, value=0, description='c'),
         levels=IntSlider(min=5, max=20, step=1, value=10, description='Levels')
        );
```

## Questions

1. As $C$ is increased, the third term in the equations ($C Q_1 Q_2^2$) begins to convey a coupling between $Q_1$ and $Q_2$. This causes the low-energy contours in the $Q_1$ vs. $Q_2$ plot to fall out of equilibrium, or lose its symmetry about the axes. As $Q_2$ is squared, its greater influence on the coupled system than $Q_1$ can be easily observed at higher values of $Q_2$.
2. Yes, the force on $Q_1$ does depend on the direction of $Q_2$ because of the same term as before, which connects the two variables. Changing the directions of $Q_2$ will affect the potential energy with respect to $Q_1$, which indicates that the force on $Q_1$ will change based on the sign of $Q_2$.
3. If $Q_2$ is driven, such as through oscillations or external force, $Q_1$ would be affected because of the term $C Q_1 Q_2^2$, which couples $Q_1$ and $Q_2$. If $Q_2$ is affected, $Q_1$ is affected.
4. If $Q_1$ is moved away from equilibrium, $Q_2$ may potentially be affected due to the  term, which couples the two variables. As $Q_1$ is out of equilibrium, the force around $Q_2$ would be asymmetric, likely leading to a change in $Q_2$.
